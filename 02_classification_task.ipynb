```python
img = img[50:250, 40:240]
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
```

```python
img_resized = cv2.resize(img_gray, (25, 25))
plt.imshow(img_resized, cmap='gray')
```
